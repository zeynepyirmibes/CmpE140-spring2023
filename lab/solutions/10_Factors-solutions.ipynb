## Question 3
Imagine you have a data about movies watched by a user of a streaming service in a month. Create a function called check_movies, that:
1. Takes a vector containing movie categories
2. Transforms it into a factor object
3. Creates a frequency table that shows the movies watched by a user in decreasing order
4. Returns the two most watched movie genres

For example if your input is:
```r
set.seed(3)

movies_watched <- sample(c("Adventure", "Sci-Fi", "Comedy", "Horror", "Drama", "Romance"), 20, replace = TRUE)
```
Your output should be:
```r
The two most frequent categories are: Drama and Sci-Fi
```

*Hint: use cat() function so that your result is shown in the required format


````r
set.seed(3)
movies_watched <- sample(c("Adventure", "Sci-Fi", "Comedy", "Horror", "Drama", "Romance"), 20, replace = TRUE)

check_movies <- function(movies_watched){
  movies_categories <- factor(movies_watched)
  freq_table <- sort(table(movies_categories), decreasing = TRUE)
  top_2 <- names(freq_table)[1:2]
  cat("The two most frequent categories are:", top_2[1], "and", top_2[2])
}

check_movies(movies_watched)
````

| days      | intensity |

| wednesday | none      |
| thursday  | intense   |

2) Create a function **weekly_ex** which takes a factor argument **f** that contains the intensity of the daily exercises that a person does in a week. Your function should find the days of the week that the individual works out (so the intensity must be "intense" or "light") and submit those days in a list's first element which has **"well_done"** as a name. Then it should find the days that the individual does not work out (so the intensity must be "none") and submit those in the list's second element that has **"shame"** as a name. Lastly, your function should replace the "none"s in the factor **f** with "light"s and submit the new factor in the last element of the list which has a name of **"next_week"** (remember doing the necessary arrangement in the levels of the factor). 

3) Try your function with **ex_fac**. Output must be as such:
```r
> weekly_ex(ex_fac)

$well_done
[1] "monday"   "tuesday"  "thursday" "friday"  

$shame
[1] "wednesday" "saturday"  "sunday"   

$next_week
[1] intense light   light   intense light   light   light  
Levels: light intense

```

````r
#1
ex_fac <- factor(c("intense", "light", "none", "intense", "light", "none", "none"))

#2
weekly_ex <- function(f)
{
  week <- c("monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday")
  well_done <- week[f != "none"]
  shame <- week[f == "none"]
  f[f == "none"] <- "light"
  f <- droplevels(f)
  thelist <- list(well_done = well_done, shame = shame, next_week = f)
  return(thelist)
}

#3
weekly_ex(ex_fac)
````

## Question 10 

Suppose you are given a data frame with two columns: one contains the type of fruit and the other contains the weight of the fruit in grams, such as:

```r
df <- data.frame(Fruit = c("Apple", "Banana", "Orange", "Apple", "Orange", "Orange", "Banana"),
                 Weight = c(50, 100, NA, 75, 85, 80, NA))
```
Apply below steps to this data frame:
1) Convert the fruit column to a factor and calculate the average weight of each fruit type. 
2) As the data contains some missing values in the weight column, replace these missing values with the average weight of the same fruit type. 
3) Finally, print the average value for each fruit type using `tapply()` function.

## Solution

```r

df <- data.frame(Fruit = c("Apple", "Banana", "Orange", "Apple", "Orange", "Orange", "Banana"),
                 Weight = c(50, 100, NA, 75, 85, 80, NA))

df$Fruit <- factor(df$Fruit)

for (fruit in levels(df$Fruit)) {
  mean_weight <- mean(df$Weight[df$Fruit == fruit], na.rm = TRUE) # calculate mean
  df$Weight[df$Fruit == fruit] <- ifelse(is.na(df$Weight[df$Fruit == fruit]), mean_weight, df$Weight[df$Fruit == fruit])
  # above line is this: if weight of that fruit is NA, assign it to be average; if it is not, then leave it as is
}

print(df)
print(tapply(df$Weight, df$Fruit, mean))


```

````r
df <- data.frame(Fruit = c("Apple", "Banana", "Orange", "Apple", "Orange", "Orange", "Banana"),
                 Weight = c(50, 100, NA, 75, 85, 80, NA))

df$Fruit <- factor(df$Fruit)

for (fruit in levels(df$Fruit)) {
  mean_weight <- mean(df$Weight[df$Fruit == fruit], na.rm = TRUE) # calculate mean
  df$Weight[df$Fruit == fruit] <- ifelse(is.na(df$Weight[df$Fruit == fruit]), mean_weight, df$Weight[df$Fruit == fruit])
  # above line is this: if weight of that fruit is NA, assign it to be average; if it is not, then leave it as is
}

print(df)
print(tapply(df$Weight, df$Fruit, mean))
````
